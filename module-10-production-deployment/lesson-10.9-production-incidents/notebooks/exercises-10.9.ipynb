## Ex 1: Severity Ladder


```python
print('Severity ladder (classify by USER IMPACT, not internal cause):')
for sev, impact, response_time in [
    ('SEV1','Total outage or data loss','Page all, status-page < 15 min'),
    ('SEV2','Major feature / tenant down','Page on-call, fix this shift'),
    ('SEV3','Degraded but usable','Ticket, address this week'),
    ('SEV4','Minor defect','Backlog, next sprint'),
]:
    print(f'  {sev:6s}: {impact:32s} | Response: {response_time}')
```

## Ex 2: On-Call Tool Comparison


```python
print('On-call / paging tools:')
for tool, strength, best_for in [
    ('PagerDuty','Richest integrations, industry std','Large teams w/ budget'),
    ('Opsgenie','Atlassian stack fit','Jira/Confluence shops'),
    ('VictorOps','Splunk-native','Splunk-centric SRE'),
    ('Squadcast','Modern UX, cheaper','India/APAC, startups'),
    ('GCP Incident Response','Native GCP hooks','GCP-heavy shops'),
]:
    print(f'  {tool:20s}: {strength:32s} | Best: {best_for}')
```

## Ex 3: Common LLM Incidents


```python
print('The 5 common LLM-specific incidents + mitigations:')
for incident, signal, mitigation in [
    ('Provider outage','5xx spike on one provider','Gateway fallback to alt'),
    ('Rate-limit storm','429 surge, cost spike','Queue + shed at front door'),
    ('Prompt injection abuse','Anomalous tool calls','Isolate tenant, re-auth, block'),
    ('Cost runaway','$/hour breach','Virtual-key kill switch'),
    ('Quality drift','Eval-on-prod score drop','Pin prior model, rollback prompt'),
]:
    print(f'  {incident:22s}: {signal:28s} | Fix: {mitigation}')
```

## Ex 4: War-Room Roles


```python
print('War-room roles (SEV1 / SEV2):')
for role, responsibility, must_not_do in [
    ('Incident Commander','Coordinate, decide, escalate','Type commands / fix the bug'),
    ('Comms Lead','Status page, stakeholders','Investigate technical cause'),
    ('Tech Lead','Execute the fix','Make coordination decisions'),
    ('Scribe (optional)','Real-time timeline','Take on another role mid-incident'),
]:
    print(f'  {role:18s}: do {responsibility:28s} | DO NOT: {must_not_do}')
```

## Ex 5: Status Page Cadence


```python
print('Status page state machine:')
for state, post_when, example_copy in [
    ('Investigating','Within 15 min of detection','"We are investigating elevated errors..."'),
    ('Identified','Root cause known','"Upstream provider degradation; failing over."'),
    ('Monitoring','Fix deployed','"Fix deployed; monitoring for stability."'),
    ('Resolved','Alerts clear + verified','"Service restored; postmortem to follow."'),
]:
    print(f'  {state:14s}: {post_when:30s} | {example_copy}')
print()
print('Short sentences. No internal jargon. Technical timeline lives in the postmortem.')
```

## Ex 6: Postmortem Sections


```python
print('Blameless postmortem required sections:')
for section, contains in [
    ('Summary','1 paragraph: what happened, what was impacted'),
    ('Impact','users/$/duration/affected systems'),
    ('Timeline','events with IST timestamps, links to Slack/PR'),
    ('Root Cause','technical cause; no blame on individuals'),
    ('Contributing Factors','latent issues that made it worse'),
    ('What Went Well','reinforce working parts'),
    ('What Went Badly','process / tooling gaps'),
    ('Action Items','SMART, owner + due + ticket'),
]:
    print(f'  {section:22s}: {contains}')
print()
print('Published within 5 business days. Reviewed in team ops meeting.')
```

## Ex 7: MTTR vs MTBF


```python
print('MTTR vs MTBF - where to invest:')
for metric, meaning, levers in [
    ('MTTR','Mean Time To Recover','Tested runbooks, version-pointer rollback, pre-warmed fallback'),
    ('MTBF','Mean Time Between Failures','Code quality, infra redundancy, chaos engineering'),
]:
    print(f'  {metric}: {meaning}')
    print(f'    levers -> {levers}')
print()
print('MTTR moves faster than MTBF. A quarter of rehearsal can halve MTTR.')
print('Users experience MTTR directly; MTBF is an internal metric.')
```
