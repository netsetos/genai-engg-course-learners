## Ex 1: Why Streaming Matters — Perceived Performance


```python
print('Perceived performance vs actual latency:')
for scenario, ttft_ms, total_ms, perceived in [
    ('Non-streaming 4s',   4000, 4000, 'feels broken'),
    ('Stream TTFT 280ms',   280, 4000, 'feels fast (~5x better)'),
    ('Stream TTFT 100ms',   100, 4000, 'feels instant'),
    ('Stream TTFT 1200ms', 1200, 4000, 'noticeable lag'),
]:
    print(f'  {scenario:28s} | TTFT={ttft_ms:>4}ms total={total_ms}ms -> {perceived}')
print()
print('Rule: stream anything > 500ms total expected duration. TTFT < 500ms target.')
```

## Ex 2: Streaming Protocols


```python
print('Streaming protocol comparison:')
for proto, direction, best_for in [
    ('SSE','Server -> Client','Default for token streaming'),
    ('WebSocket','Bidirectional','Mid-stream cancel, voice, tool approval'),
    ('Long-poll','Client polls','Fallback when SSE blocked by proxy'),
    ('HTTP/2 push','Server push','Rare; SSE easier in practice'),
    ('WebTransport','Bidirectional, QUIC','Early; good for multimodal'),
]:
    print(f'  {proto:14s}: {direction:18s} | Best: {best_for}')
```

## Ex 3: UX Patterns That Move the Needle


```python
print('UX patterns for LLM UIs:')
for pattern, purpose, tip in [
    ('Skeleton / shimmer','Signal activity before first token','Swap on first byte, not first word'),
    ('Optimistic UI','Commit intent immediately','Rollback on failure; never half state'),
    ('Progressive disclosure','Show CoT / tool calls on demand','Collapse by default'),
    ('Streaming','Render tokens as received','TTFT < 500ms; 5x perceived speedup'),
    ('Tool-call pills','Name what is happening','"Searching docs..." > generic spinner'),
]:
    print(f'  {pattern:22s}: {purpose:38s} | Tip: {tip}')
```

## Ex 4: Error States Library


```python
print('The 5 LLM UI error states (each with a concrete next action):')
for state, trigger, action in [
    ('rate_limit','Upstream 429','Countdown + auto-retry'),
    ('timeout','Client timeout','Retry button'),
    ('refusal','Model refuses','Show reason; suggest alt phrasing'),
    ('outage','Provider down','Banner: switched to backup, quality may vary'),
    ('cost_cap','Budget exceeded','Upgrade link, alt model suggestion'),
]:
    print(f'  {state:12s}: trigger={trigger:22s} | action={action}')
print()
print('Never show a generic "error". Every state has a concrete path forward.')
```

## Ex 5: Feedback Signals


```python
print('Feedback signals and what they feed:')
for signal, value, feeds_into in [
    ('Thumbs up/down','Fast binary quality','Eval-on-prod, prompt iteration'),
    ('Edit / correction','Ground truth delta','Fine-tune dataset, RAG gap analysis'),
    ('Regenerate','Implicit dissatisfaction','Low-confidence example flag'),
    ('Star / save','Gold examples','Few-shot library, test set growth'),
    ('Copy','Used the answer','Lightweight success proxy'),
]:
    print(f'  {signal:18s}: {value:30s} | Feeds: {feeds_into}')
```

## Ex 6: Accessibility Requirements


```python
print('Accessibility checklist for streaming LLM UIs:')
for req, how, tool in [
    ('Screen reader announcements','aria-live="polite" debounced by sentence','NVDA/VoiceOver'),
    ('Keyboard send/cancel','Enter / Shift+Enter / Esc','manual test'),
    ('Reduced motion','Honor prefers-reduced-motion','CSS @media'),
    ('Color contrast','WCAG 2.2 AA (4.5:1 body, 3:1 UI)','axe-core'),
    ('Focus management','Return focus to input post-send','manual + Lighthouse'),
]:
    print(f'  {req:28s}: {how:42s} | Tool: {tool}')
```

## Ex 7: Citation Rendering


```python
print('Citation UX contract (backend + frontend):')
for piece, contract, ux_effect in [
    ('source_id on each chunk','UUID per source doc','Stable anchor for jump'),
    ('char_offset start/end','Integer range','Highlight on click'),
    ('confidence (optional)','0..1','Color badge (green/amber/red)'),
    ('url + title','Linkable source','Hover card + source panel link'),
    ('snippet','<= 240 chars','Hover preview w/o leaving answer'),
]:
    print(f'  {piece:24s}: {contract:22s} | UX: {ux_effect}')
```
