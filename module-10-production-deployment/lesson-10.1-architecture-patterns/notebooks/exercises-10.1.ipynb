## Architecture Decision Matrix


```python
for arch,when,limit in [
    ('Modular Monolith','API-calling, MVPs, <10 eng','GPU/CPU cant scale independently'),
    ('Microservices','Multi-model, polyglot, GPU scaling','Network latency, complexity'),
    ('Serverless','Event-driven, bursty','No GPU (Lambda), cold starts'),
    ('Hybrid (consensus)','Monolith core + GPU microservice','Requires clear module boundaries'),
]:
    print(f'  {arch:20s}: {when:45s} | Limit: {limit}')
```

## API Protocol Comparison


```python
for proto,use,perf in [
    ('REST','Public APIs, browser clients','93% adoption, standard'),
    ('gRPC','Internal service-to-service','10x lower latency, 19% less CPU'),
    ('GraphQL','Variable client data shapes','Flexible queries'),
    ('SSE','LLM token streaming','Industry standard, auto-reconnect'),
]:
    print(f'  {proto:10s}: {use:35s} | {perf}')
```

## Vector DB Selection


```python
for db,strength,best_for in [
    ('Qdrant','p50<5ms, Rust, 32x compression','Performance + filtering'),
    ('Pinecone','Zero-ops, SOC2/HIPAA','Fastest path to production'),
    ('pgvector','5.7x improvement, same DB','Teams on PostgreSQL, <50M vectors'),
    ('ChromaDB','Zero-config Python','Prototyping, <10M vectors'),
]:
    print(f'  {db:12s}: {strength:35s} | Best: {best_for}')
```
