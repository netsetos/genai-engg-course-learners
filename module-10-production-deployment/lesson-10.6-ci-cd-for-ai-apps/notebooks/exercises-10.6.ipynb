## Ex 1: Why AI CI/CD Differs


```python
print('AI-specific CI/CD concerns:')
for concern, impact, mitigation in [
    ('Non-determinism','Same prompt gives different output','Eval gates, not exact match'),
    ('Paid APIs per run','CI cost spikes','Virtual key caps, sampling'),
    ('Prompt-as-code','Silent quality drift','Git + semver prompts'),
    ('Model updates','Provider swaps model silently','Pin by version when possible'),
    ('Long-tail failures','95% pass masks edge cases','Regression set + diverse evals'),
]:
    print(f'  {concern:20s}: {impact:38s} | {mitigation}')
```

## Ex 2: CI Tool Landscape


```python
print('CI tool comparison:')
for tool, strength, best_for in [
    ('GitHub Actions','Tight GH integration, marketplace','Most teams default'),
    ('GitLab CI','Built-in everything, self-host','Self-managed full stack'),
    ('CircleCI','Fast Docker cache, mature','Polyglot teams'),
    ('Jenkins','Max flexibility, plugins','Legacy / enterprise'),
    ('Dagger','Portable pipelines as code','Multi-CI future-proof'),
]:
    print(f'  {tool:18s}: {strength:36s} | Best: {best_for}')
```

## Ex 3: Eval Gate Types


```python
print('Eval gate options:')
for gate, measures, cost in [
    ('LLM-judge','Holistic quality (Claude/GPT-4 grader)','~$0.01/sample'),
    ('Exact match','Deterministic answers (math, code)','free'),
    ('Embedding similarity','Semantic equivalence to gold','~$0.0001/sample'),
    ('Regex / schema','Structured output compliance','free'),
    ('Toxicity / PII','Safety regression','low (local model)'),
    ('Pairwise preference','A/B against baseline','~$0.02/pair'),
]:
    print(f'  {gate:24s}: {measures:40s} | Cost: {cost}')
```

## Ex 4: Deploy Strategies


```python
print('Deploy strategies:')
for strat, risk, when in [
    ('Blue/green','Atomic 0->100, easy revert','Stateless apps, quick rollback critical'),
    ('Canary 5/50/100','Gradual, auto-rollback','Default for LLM apps'),
    ('Rolling','Slow, no extra cost','Pod-by-pod K8s default'),
    ('Shadow','Zero user risk, compare offline','Model / prompt swaps before promote'),
    ('Feature flag','Instant toggle, A/B','Prompt version rollout'),
]:
    print(f'  {strat:16s}: {risk:32s} | When: {when}')
```

## Ex 5: Canary Progression


```python
print('Typical canary timeline:')
for t_min, traffic_pct, action in [
    (0,   5, 'start canary 5%'),
    (5,   5, 'quality check: pass -> continue'),
    (30,  5, 'confirmed stable'),
    (30, 50, 'promote to 50%'),
    (60, 50, 'quality + error check: pass'),
    (60,100, 'promote to 100%'),
    (90,100, 'bake; keep prior revision warm for 24h'),
]:
    print(f'  t+{t_min:3d}min | {traffic_pct:3d}% traffic | {action}')
```

## Ex 6: Secret Management (OIDC > Static)


```python
print('Secret management comparison:')
for approach, leak_risk, rotation_cost in [
    ('Hardcoded in repo','CRITICAL', 'impossible without rewrite'),
    ('GH Actions Secret','Low','Manual update per secret'),
    ('OIDC -> GCP SM','Very low','Rotate in UI; no code change'),
    ('OIDC -> AWS STS','Very low','Rotate in Secrets Manager'),
    ('HashiCorp Vault','Low','Centralized, more ops overhead'),
]:
    print(f'  {approach:22s}: leak={leak_risk:10s} | rotate={rotation_cost}')
```

## Ex 7: Rollback Time Budget


```python
print('Target MTTR by rollback path:')
for path, target_mttr, notes in [
    ('Version-pointer flip (Redis)','< 30 s','Instant; no deploy'),
    ('Cloud Run traffic split','< 90 s','gcloud run update-traffic'),
    ('Git revert + canary deploy','< 20 min','Through full pipeline'),
    ('Full rebuild + redeploy','< 60 min','Worst case, avoid'),
]:
    print(f'  {path:34s}: MTTR {target_mttr:7s} | {notes}')
print()
print('Rehearse rollback monthly or it is not a rollback.')
```

## Recap
Treat prompts as code: Git + semver + eval gate on every PR.
Canary 5/50/100 default; shadow for model swaps. Auto-rollback on quality regression.
OIDC > static keys. Virtual-key cost caps per env.
Rehearse rollback monthly; target < 90s via version-pointer flip.
