## Ex 1: Agent = LLM + Tools + Memory + Autonomy


```python
print('Agent Components:')
for comp, role in [
    ('LLM','Brain — reasoning, planning, decision-making'),
    ('Tools','Hands — interact with external systems'),
    ('Memory','Context — short-term + long-term knowledge'),
    ('Autonomy','Agency — LLM controls workflow execution'),
]: print(f'  {comp:10s}: {role}')
print()
print('Taxonomy: Chatbot → Copilot → Agent → Autonomous')
```

## Ex 2: The Agent Loop


```python
print('Core agent loop (pseudocode):')
print()
print('messages = [system_prompt + user_task]')
print('while True:')
print('    response = llm.chat(messages, tools=TOOLS)')
print('    if response.stop_reason != "tool_use":')
print('        break  # Task complete')
print('    for tool_call in response.tool_calls:')
print('        result = execute(tool_call)')
print('        messages.append(tool_call)')
print('        messages.append(tool_result(result))')
print()
print('That\'s it. This drives Claude Code, OpenAI SDK, everything.')
```

## Ex 3: Error Compounding


```python
print('Error compounding across agent steps:')
print()
for steps in [1, 3, 5, 10, 15, 20, 25]:
    success = 0.95 ** steps
    print(f'  {steps:2d} steps @ 95%/step = {success*100:.1f}% overall')
print()
print('Key insight: even 95% per step = 60% at 10 steps')
print('This is why max_turns caps of 15-25 are essential')
```

## Ex 4: Token Cost Compounding


```python
import math
S = 1000  # system prompt
u = 2500  # new input per iteration
r = 500   # output per iteration
print('Token cost growth (N iterations):')
print(f'  System prompt: {S} tokens')
print(f'  New input/iter: {u} tokens')
print(f'  Output/iter: {r} tokens')
print()
for N in [1, 3, 5, 10, 15, 20]:
    total = N*S + u*N*(N+1)//2 + r*N*(N-1)//2
    cost = total * 3 / 1_000_000  # Claude Sonnet input pricing
    single = (S + u) * 3 / 1_000_000
    mult = cost / single if single > 0 else 0
    print(f'  N={N:2d}: {total:>8,} tokens = ${cost:.3f} ({mult:.0f}× single call)')
```
