## Ex 1: Anthropic's 6 Single-Agent Patterns


```python
print('Anthropic\'s building blocks (simplest → most complex):')
for i,(p,desc,use) in enumerate([
    ('Augmented LLM','Retrieval + Tools + Memory','Default. Most apps need nothing more'),
    ('Prompt Chaining','Fixed sequence + gates','Known decomposition, trade latency for accuracy'),
    ('Routing','Classify → specialize','Known categories, model/prompt optimization'),
    ('Parallelization','Sectioning or voting','Independent subtasks or confidence boost'),
    ('Orchestrator-Workers','Dynamic delegation','Unknown subtasks at runtime'),
    ('Evaluator-Optimizer','Generate → evaluate → iterate','Iterative quality improvement'),
],1): print(f'  {i}. {p:25s}: {desc}')
   print(f'     Use: {use}')
```

## Ex 3: Handoff Strategies Compared


```python
print('Framework handoff comparison:')
for fw, ctx, best in [
    ('OpenAI SDK','Full conversation history (default)','Conversational routing'),
    ('Claude Code','Clean-slate (fresh context, prompt only)','Independent parallel tasks'),
    ('Google ADK','Shared session state (output_key)','Deterministic pipelines'),
]: print(f'  {fw:12s}: {ctx}')
   print(f'  {"":12s}  Best for: {best}\n')
```

## Ex 4: LangGraph State Schema


```python
print('LangGraph TypedDict with reducers:')
print()
print('class AgentState(TypedDict):')
print('    messages: Annotated[list, add_messages]  # Smart append')
print('    results: Annotated[list, operator.add]   # Always concat')
print('    step: str                                 # Overwrite')
print()
print('Reducer semantics:')
print('  add_messages → append with ID dedup')
print('  operator.add → always concatenate')
print('  no annotation → last writer wins (overwrite)')
```

## Ex 5: A2A vs MCP


```python
print('A2A vs MCP — complementary protocols:')
print()
for dim, mcp, a2a in [
    ('Purpose','Agent ↔ Tools','Agent ↔ Agent'),
    ('Analogy','USB-C of AI','HTTP of agents'),
    ('Scope','Single agent capabilities','Cross-org communication'),
    ('Discovery','Tool schemas','Agent Cards (JSON)'),
    ('Work unit','Tool call + result','Task lifecycle'),
    ('Status','Mature (spec 2025-06-18)','v1.0 (150+ orgs)'),
]: print(f'  {dim:12s}: MCP={mcp:30s} | A2A={a2a}')
```
