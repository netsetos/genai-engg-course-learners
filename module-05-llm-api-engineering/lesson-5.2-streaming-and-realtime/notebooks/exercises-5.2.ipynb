## Ex 1: SSE Protocol

## Ex 2: TTFB Comparison


```python
providers = [
    ('Groq (LPU)', '207ms', '241-840', 'Custom ASIC silicon'),
    ('OpenAI GPT-4o', '~500ms', '50-100', 'GPU inference'),
    ('Google Gemini', '~1.0s', '40-60', 'TPU inference'),
    ('Anthropic Claude', '~2.0s', '28-35', 'GPU inference'),
    ('DeepSeek V3', '7-19s', '25-32', 'MoE, variable load'),
]
print(f"{'Provider':22s} {'TTFB':>8s} {'tok/s':>10s} {'Notes'}")
print('-'*60)
for name, ttfb, speed, notes in providers:
    print(f"{name:22s} {ttfb:>8s} {speed:>10s}  {notes}")
print()
print('Sub-500ms TTFB feels instant for interactive apps')
print('Reasoning models add seconds-to-minutes before visible tokens')
```

## Ex 3: Provider Event Formats


```python
print('OpenAI streaming events:')
print('  First:  {"choices": [{"delta": {"role": "assistant"}}]}')
print('  Middle: {"choices": [{"delta": {"content": "token"}}]}')
print('  Last:   {"choices": [{"delta": {}, "finish_reason": "stop"}]}')
print('  Usage:  opt-in via stream_options: {include_usage: true}')
print()
print('Anthropic streaming events (7 types):')
print('  message_start → content_block_start → ping →')
print('  content_block_delta → content_block_stop →')
print('  message_delta → message_stop')
print('  Input tokens in message_start, output in message_delta')
print()
print('Gemini streaming:')
print('  Multi-sentence chunks (not individual tokens)')
print('  Each chunk = complete GenerateContentResponse')
print('  Cumulative usageMetadata in every chunk')
```

## Ex 5: Production Hardening


```python
print('Reverse Proxy Configuration (streaming killers):')
print()
proxies = [
    ('nginx', 'proxy_buffering off; proxy_http_version 1.1;', 'X-Accel-Buffering: no'),
    ('Cloudflare', 'Buffers ~100KB (Enterprise to disable)', 'Use Workers or bypass'),
    ('AWS API GW', 'responseTransferMode: STREAM (Nov 2025)', '15-min timeout support'),
    ('Heartbeat', ': heartbeat every 15s', 'Prevents proxy timeouts'),
]
for proxy, fix, note in proxies:
    print(f'  {proxy:12s}: {fix}')
    print(f'               → {note}')
print()
print('Observability: log start metadata, completion via onFinish')
print('Rate limit: max 3 concurrent streams/user + token budget/hr')
```

## Ex 6: Advanced Patterns


```python
print('Advanced Streaming Patterns:')
print()
print('1. RAG Streaming:')
print('   event: status → "Searching docs..."')
print('   event: metadata → sources list')
print('   event: token → streamed answer')
print()
print('2. Tool Call Streaming:')
print('   Accumulate JSON fragments per tool_calls[index]')
print('   Parse on finish_reason: "tool_calls"')
print('   Execute → new streaming request with results')
print()
print('3. Multi-Model Racing:')
print('   asyncio.wait(tasks, return_when=FIRST_COMPLETED)')
print('   Cancel slower → return fastest')
print('   Doubles cost — fallback pattern is cheaper')
print()
print('4. Streaming Guardrails:')
print('   Buffer ~20-50 tokens → moderate → release')
print('   Adds ~500ms delay but catches harmful content')
```

## Recap
SSE is the protocol (fetch+ReadableStream, not EventSource). TTFB varies 100× (Groq 207ms vs DeepSeek 19s). Providers stream differently: OpenAI=token deltas, Anthropic=7-event lifecycle, Gemini=multi-sentence chunks. Frontend: Vercel AI SDK useChat, streaming-markdown, AbortController. Production: nginx proxy_buffering off, Cloudflare Workers, heartbeats, stream-aware observability. Advanced: RAG streaming, tool call accumulation, multi-model racing, streaming guardrails. Audio: OpenAI Realtime (WebRTC), Gemini Live. India: Mumbai endpoints eliminate 250ms, WhatsApp can't stream, Telegram can.
