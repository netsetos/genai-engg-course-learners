## Ex 3: Cost Optimization Stacking


```python
print('Cost optimization techniques (multiplicative!):')
print()
techniques = [
    ('Base price (Claude Sonnet)', '$3.00/M', '1.00×'),
    ('+ Prompt caching (90% off reads)', '$0.30/M', '0.10×'),
    ('+ Batch API (50% off)', '$0.15/M', '0.05×'),
    ('+ Tiered routing (70% to cheap model)', '~$0.50/M avg', '~0.17×'),
]
for tech, price, mult in techniques:
    print(f'  {tech:45s} {price:>10s} ({mult})')
print()
print('Combined: up to 95% reduction!')
print('Rule: any non-real-time task → batch API')
print('Rule: any repeated context → prompt caching')
```

## Ex 5: Error Handling Stack


```python
print('Production Resilience Stack:')
print()
stack = [
    ('1. Hard timeout', '30 seconds on ALL calls', 'Prevents resource exhaustion'),
    ('2. Exponential backoff', '2s→4s→8s→30s cap', 'Handles transient errors (429, 500-504)'),
    ('3. Circuit breaker', 'Trip after 5 failures', '60s cooldown, single probe to recover'),
    ('4. Multi-provider fallback', 'GPT-4o→Claude→Gemini', 'Transparent provider switching'),
    ('5. Alerting', '>5% error rate warning', '>15% critical, page on-call'),
]
for layer, config, purpose in stack:
    print(f'  {layer:25s} {config:25s} {purpose}')
```

## Ex 6: Tiered Routing Calculator


```python
daily_tokens = 10_000_000  # 10M tokens/day

# Without routing: all to GPT-4o
cost_single = daily_tokens / 1e6 * (2.50 + 10.00) / 2  # avg in+out

# With tiered routing
tiers = [
    ('Simple (70%)', 0.70, 0.27),  # DeepSeek V3
    ('Medium (25%)', 0.25, 0.375),  # GPT-4o-mini avg
    ('Complex (5%)', 0.05, 9.00),  # Claude Sonnet avg
]
cost_tiered = sum(daily_tokens * pct / 1e6 * price for _, pct, price in tiers)

print(f'10M tokens/day cost comparison:')
print(f'  Single model (GPT-4o): ${cost_single:.2f}/day = ${cost_single*30:.0f}/month')
print(f'  Tiered routing:        ${cost_tiered:.2f}/day = ${cost_tiered*30:.0f}/month')
print(f'  Savings: {(1-cost_tiered/cost_single)*100:.0f}%')
```
