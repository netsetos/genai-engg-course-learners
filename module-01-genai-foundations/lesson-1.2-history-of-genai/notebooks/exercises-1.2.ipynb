```python
import numpy as np
np.random.seed(42)
real_mean, real_std = 5.0, 1.0
gen_mean, gen_std = 0.0, 3.0
lr = 0.1

print(f"{'Epoch':>6} {'Mean':>8} {'Std':>8} {'Dist':>8}")
for epoch in range(20):
    mean_err = real_mean - gen_mean
    std_err = real_std - gen_std
    gen_mean += lr * mean_err + np.random.normal(0, 0.05)
    gen_std += lr * std_err * 0.5 + np.random.normal(0, 0.02)
    gen_std = max(0.1, gen_std)
    dist = abs(mean_err) + abs(std_err)
    if epoch % 2 == 0:
        print(f'{epoch:>6} {gen_mean:>8.2f} {gen_std:>8.2f} {dist:>8.2f}  {chr(9608)*int(dist*3)}')
```

<details><summary>Solution</summary>



```python
import numpy as np
np.random.seed(42)
cluster_0 = np.random.normal([2,2], 0.5, (50,2))

def encode(x): return x*0.1, np.zeros_like(x)-1
def reparam(mu, lv):
    return mu + np.exp(0.5*lv) * np.random.normal(0,1,mu.shape)
def decode(z): return z*10

mu, lv = encode(cluster_0[:3])
z = reparam(mu, lv)
print('Original:', cluster_0[:3].round(1))
print('Latent:', z.round(2))
print('Reconstructed:', decode(z).round(1))

print('\nInterpolation (digit 0 -> digit 1):')
for a in np.linspace(0,1,6):
    z_i = (1-a)*np.array([0.2,0.2]) + a*np.array([0.8,0.8])
    print(f'  alpha={a:.1f}: {decode(z_i).round(1)}')
```
