## Ex 1: ReAct Loop


```python
print('ReAct: Thought → Action → Observation → loop')
print()
print('Step 1: Thought — "I need the weather in Mumbai"')
print('Step 2: Action — get_weather(location="Mumbai")')
print('Step 3: Observation — {"temp": 35, "conditions": "sunny"}')
print('Step 4: Thought — "I have the weather. Generate response."')
print('Step 5: Final Answer — "It\'s 35°C and sunny in Mumbai."')
print()
print('Budget limits for production:')
for limit, value in [
    ('Max steps','25'),
    ('Max tool calls','12'),
    ('Max wall-clock time','60 seconds'),
    ('Max cost','$1.00 per run'),
    ('Loop detection','Hash(tool+args), terminate on repeat'),
]: print(f'  {limit:25s}: {value}')
```

## Ex 2: Architecture Decision Tree


```python
print('Agent architecture decision tree:')
print()
for tools, pattern, framework in [
    ('3-10 tools, single domain','Single-agent ReAct','Any framework'),
    ('10-20 tools, 2-3 domains','Router → specialists','OpenAI Agents SDK'),
    ('Complex with human approval','Stateful workflow','LangGraph'),
    ('Team-based collaboration','Role-based agents','CrewAI'),
    ('50+ agents, enterprise','Hierarchical delegation','LangGraph + CrewAI'),
    ('Cross-framework agents','Agent-to-agent protocol','A2A'),
]: print(f'  {tools:30s} → {pattern:25s} → {framework}')
```

## Ex 3: Saga Pattern


```python
print('Saga pattern for multi-tool error handling:')
print()
steps = [
    ('book_flight','cancel_flight','Flight booked: AI-123'),
    ('book_hotel','cancel_hotel','FAILED: No rooms available'),
    ('rent_car','cancel_car','(never reached)'),
]
print('Forward execution:')
for i,(fwd,comp,result) in enumerate(steps,1):
    status = '❌' if 'FAILED' in result else '✅'
    print(f'  Step {i}: {fwd}() → {status} {result}')
print()
print('Compensation (reverse order):')
print('  Step 2: hotel failed → no compensation needed')
print('  Step 1: cancel_flight() → ✅ Flight AI-123 cancelled')
print()
print('Result: User NOT charged for orphaned flight')
```

## Ex 4: Token Budget Management


```python
print('Token budget compounding in multi-agent systems:')
print()
for scenario, tokens in [
    ('Single request (uncompressed)','~10,500 tokens'),
    ('After aggressive compression','~650 tokens (94% reduction)'),
    ('3-agent pipeline × 1 cycle','~6,000 tokens accumulated'),
    ('3-agent pipeline × 10 cycles','~60,000 tokens accumulated'),
]: print(f'  {scenario:40s}: {tokens}')
print()
print('Compaction strategy:')
print('  70% utilization → warning log')
print('  85% utilization → summarize oldest 20 turns')
print('  90% utilization → aggressive compaction (60-80% reduction)')
print('  Keep last 3 turns raw for model momentum')
print('  Store exact messages externally for audit/replay')
```

## Recap
ReAct is the dominant agent architecture — Thought→Action(function call)→Observation→loop. Add budget limits from day 1 (max steps, calls, time, cost). Choose architecture by complexity: single-agent→router→planner→hierarchical. State management is the hardest problem — compact at 70/85/90%. Use Saga pattern for tools that modify external state. Framework choice: LangGraph for complex stateful, OpenAI SDK for quick prototypes, CrewAI for team-based. Claude Code proves simple loops outperform complex swarms. MCP for tools, A2A for agent communication. India: Sarvam full-stack + E2E Networks + Bhashini = fraction of global cost.
