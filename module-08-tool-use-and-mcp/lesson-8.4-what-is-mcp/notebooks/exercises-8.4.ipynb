## Ex 1: The N×M Problem


```python
print('Without MCP: M apps × N tools = M×N integrations')
print('With MCP: M clients + N servers = M+N integrations')
print()
for m,n in [(3,5),(5,10),(10,50)]:
    print(f'  {m} apps × {n} tools: Without={m*n}, With MCP={m+n}, Savings={m*n-m-n}')
```

## Ex 2: MCP Lifecycle


```python
print('MCP Lifecycle:')
print()
for phase, method, detail in [
    ('1. Initialize','initialize','protocolVersion + capabilities + clientInfo'),
    ('','notifications/initialized','Client confirms readiness'),
    ('2. Discover','tools/list','Paginated: cursor/nextCursor'),
    ('','resources/list','URI-based read-only data'),
    ('','prompts/list','User-controlled templates'),
    ('3. Execute','tools/call','name + arguments → content[] + structuredContent'),
    ('','resources/read','URI → text or base64 blob'),
    ('','prompts/get','name + args → PromptMessage[]'),
    ('4. Shutdown','(transport-level)','Close stdin or HTTP connection'),
]: print(f'  {phase:15s} {method:30s} {detail}')
```

## Ex 3: Six Primitives


```python
print('MCP Six Primitives:')
print()
for prim, ctrl, direction, purpose in [
    ('Tools','Model','Server→','Functions LLM decides to call'),
    ('Resources','App/Host','Server→','Read-only data by URI'),
    ('Prompts','User','Server→','Message templates (slash commands)'),
    ('Sampling','Server','←Client','Server requests LLM completion'),
    ('Roots','Client','←Client','Filesystem scope boundaries'),
    ('Elicitation','Server','←Client','Server asks user questions (2025-06-18)'),
]: print(f'  {prim:15s} | {ctrl:10s} | {direction:10s} | {purpose}')
```

## Ex 5: FastMCP Server


```python
print('FastMCP — the dominant MCP server framework (~70%)')
print()
print('from fastmcp import FastMCP')
print()
print('mcp = FastMCP("Weather Server")')
print()
print('@mcp.tool()')
print('def get_weather(city: str, units: str = "celsius") -> dict:')
print('    """Get current weather for a city."""')
print('    return {"temperature": 22.5, "conditions": "Partly cloudy"}')
print()
print('@mcp.resource("config://settings")')
print('def get_settings() -> str:')
print('    """Server configuration."""')
print('    return \'{"theme": "dark"}\'')
print()
print('@mcp.prompt()')
print('def debug_error(error_message: str) -> str:')
print('    return f"Debug this: {error_message}"')
print()
print('# Function name → tool name')
print('# Docstring → description')
print('# Type hints → JSON Schema')
```

## Ex 6: Security — OWASP MCP Top 10


```python
print('OWASP MCP Top 10 (v0.1 beta):')
print()
for code_id, name, risk in [
    ('MCP01','Token Mismanagement','Hard-coded creds, long-lived tokens'),
    ('MCP03','Tool Poisoning','Hidden instructions in descriptions (84.2% success)'),
    ('MCP05','Command Injection','LLM-generated input in system commands'),
    ('MCP06','Intent Flow Subversion','Context hijacks agent goal'),
    ('MCP09','Shadow MCP Servers','Unapproved servers outside governance'),
]: print(f'  {code_id}: {name}')
   print(f'    Risk: {risk}')
   print()
print('Average server security score: 34/100')
print('30+ CVEs filed in early 2026')
print('Defense: mcp-scan (hash pinning), never auto-approve')
```

## Ex 7: MCP vs Function Calling vs A2A


```python
print('Three protocols, three purposes:')
print()
for proto, purpose, when in [
    ('Function Calling','Per-request tool schemas','Simple, single-provider'),
    ('MCP','Cross-vendor tool standardization','Portable, ecosystem reuse'),
    ('A2A','Agent-to-agent communication','Multi-agent collaboration'),
]: print(f'  {proto:20s}: {purpose}')
   print(f'  {"":20s}  When: {when}')
   print()
print('MCP = vertical (agent→tool)')
print('A2A = horizontal (agent→agent)')
print('Complementary, not competing')
print('Both under Agentic AI Foundation (Linux Foundation)')
```

## Recap
MCP solves the N×M integration problem — one universal protocol for AI tool connections. Six primitives (tools/resources/prompts + sampling/roots/elicitation) with three-way control (model/app/user). FastMCP @decorator pattern for 10-line servers (~70% market share). VS Code Copilot is most spec-complete client (7/7). Security is the weakest link (84.2% tool poisoning, 34/100 avg score). MCP + A2A = complete agent infrastructure (vertical + horizontal). India: MoSPI government server exists, Bhashini MCP is the biggest opportunity.
