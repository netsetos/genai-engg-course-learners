## Ex 1: Stateless vs Stateful


```python
print('Stateless vs Stateful MCP on Cloud Run:')
for mode, flag, scaling, features in [
    ('Stateless','stateless_http=True','Horizontal (no session affinity)','No sampling/elicitation'),
    ('Stateful','default','Single instance or sticky sessions','Full MCP features'),
]: print(f'  {mode:10s}: {flag}')
   print(f'  {"":10s}  Scaling: {scaling}')
   print(f'  {"":10s}  Tradeoff: {features}')
```

## Ex 2: Optimized Dockerfile


```python
print('Production Dockerfile pattern:')
print()
print('FROM python:3.13-slim AS build')
print('COPY --from=ghcr.io/astral-sh/uv:0.8.21 /uv /bin/')
print('ENV UV_COMPILE_BYTECODE=1 UV_LINK_MODE=copy')
print('COPY uv.lock pyproject.toml ./')
print('RUN uv sync --no-install-project --no-dev  # cached layer')
print('COPY . .')
print('RUN uv sync --frozen --no-dev')
print()
print('FROM python:3.13-slim AS runtime')
print('RUN useradd -u 1001 appuser')
print('USER appuser')
print('ENTRYPOINT ["python", "server.py"]')
```

## Ex 4: OAuth 2.1 + PKCE


```python
print('OAuth 2.1 flow for MCP:')
for step, action in [
    ('1','Client connects → 401 Unauthorized'),
    ('2','Discover /.well-known/oauth-protected-resource'),
    ('3','Fetch AS metadata (RFC 8414)'),
    ('4','Dynamic Client Registration or CIMD'),
    ('5','Authorize with PKCE (S256 mandatory)'),
    ('6','Exchange code + code_verifier → access token'),
    ('7','Bearer token on all MCP requests'),
]: print(f'  Step {step}: {action}')
print()
print('Token isolation: NEVER send server A token to server B')
print('IAM for service-to-service, OAuth for external clients')
```

## Ex 5: CI/CD Pipeline


```python
print('GitHub Actions → Cloud Run pipeline:')
print()
for step, detail in [
    ('Auth','Workload Identity Federation (no JSON keys)'),
    ('Build','Docker → Artifact Registry ($PROJECT/mcp-repo/server:$SHA)'),
    ('Test','In-memory Client(server) + Inspector CLI'),
    ('Deploy','gcloud run deploy --no-traffic (canary)'),
    ('Validate','Traffic split: 10% → monitor → 100%'),
    ('Rollback','Tagged revisions for instant rollback'),
]: print(f'  {step:10s}: {detail}')
```
