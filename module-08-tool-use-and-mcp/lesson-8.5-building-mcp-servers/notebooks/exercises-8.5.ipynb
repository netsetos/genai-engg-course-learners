## Ex 1: FastMCP Decorator Pattern


```python
print('FastMCP decorator → MCP primitive:')
print()
print('@mcp.tool')
print('def add(a: int, b: int) -> int:')
print('    """Add two numbers."""')
print('    return a + b')
print()
print('Auto-generates:')
print('  name: "add"')
print('  description: "Add two numbers."')
print('  inputSchema: {type: object, properties: {a: {type: integer}, b: {type: integer}}}')
print('  outputSchema: {type: object, properties: {result: {type: integer}}}')
```

## Ex 2: Pydantic Structured Output


```python
print('Pydantic BaseModel → outputSchema + structuredContent:')
print()
print('class WeatherResult(BaseModel):')
print('    temperature: float')
print('    conditions: str')
print('    humidity: int')
print()
print('@mcp.tool')
print('def get_weather(city: str) -> WeatherResult:')
print('    return WeatherResult(temperature=22.5, conditions="sunny", humidity=65)')
print()
print('Response includes:')
print('  structuredContent: {temperature: 22.5, conditions: sunny, humidity: 65}')
print('  content: [{type: text, text: "22.5°C, sunny, 65% humidity"}]')
```

## Ex 3: Context Object


```python
print('Context API:')
for method, desc in [
    ('ctx.info(msg)','Structured log (not stdout!)'),
    ('ctx.warning(msg)','Warning-level log'),
    ('ctx.error(msg)','Error-level log'),
    ('ctx.report_progress(i, total)','Progress bar for long tasks'),
    ('ctx.read_resource(uri)','Access other resources'),
    ('ctx.sample(prompt)','Request LLM completion'),
    ('ctx.set_state(k, v)','Session-scoped state (async in v3)'),
    ('ctx.get_state(k)','Retrieve session state'),
]: print(f'  {method:35s} → {desc}')
print()
print('RULE: Never print() in stdio servers!')
print('Use ctx.info() or print(msg, file=sys.stderr)')
```

## Ex 4: Server Composition


```python
print('FastMCP mount() composition:')
print()
print('weather = FastMCP("Weather")')
print('calendar = FastMCP("Calendar")')
print()
print('main = FastMCP("MainApp")')
print('main.mount(weather, namespace="weather")')
print('main.mount(calendar, namespace="calendar")')
print()
print('Result: weather_get_forecast, calendar_list_events')
print()
print('Advanced patterns:')
print('  mount(create_proxy(url))  — remote server bridging')
print('  from_openapi(spec, client) — REST→MCP conversion')
print('  Middleware class           — audit logging, rate limiting')
print('  @mcp.tool(auth=...)       — per-component authorization')
```

## Recap
FastMCP turns Python functions into MCP servers with @decorator pattern (~70% market share). Type hints→JSON Schema, Pydantic→outputSchema.
