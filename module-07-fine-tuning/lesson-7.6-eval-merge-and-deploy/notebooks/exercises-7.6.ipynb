## Ex 2: LLM-as-Judge with Bias Mitigation


```python
print('LLM-as-Judge bias mitigation:')
print()
print('Step 1: Position Swapping')
print('  Run evaluation twice with candidate order reversed')
print('  Accept judgment ONLY when both orderings agree')
print('  Eliminates: position bias')
print()
print('Step 2: Ensemble Judging')
print('  Use GPT-4 + Claude + Gemini as judges')
print('  Majority vote (2/3 agreement required)')
print('  Eliminates: self-enhancement bias')
print()
print('Step 3: Calibration')
print('  Validate on labeled dataset (≥30 samples, ≥3 annotators)')
print('  Measure agreement before trusting on new tasks')
print('  Eliminates: systematic bias in specific domains')
print()
print('Cost comparison:')
for method, cost in [('Human eval (1000 samples)','$5,000'),
    ('GPT-4o (1000 samples)','~$50'),
    ('Prometheus self-hosted (1000)','~$30')]:
    print(f'  {method:35s}: {cost}')
```

## Ex 3: Safety Regression Testing


```python
print('Pre-deployment safety checklist:')
print()
for test, what, threshold in [
    ('HarmBench','400+ harmful behaviors, 7 categories','Refusal rate >95%'),
    ('ToxiGen','274K statements, implicit hate speech','Toxicity score <0.1'),
    ('BBQ','58K bias questions, 13 categories','Bias score <0.05'),
    ('Perplexity (WikiText-2)','General text understanding','<10% increase from base'),
    ('MMLU subset','General knowledge retention','<3% drop from base'),
    ('CCR (production)','User correction rate','<15% correction rate'),
]: print(f'  {test:25s}: {what}')
   print(f'  {"":25s}  Threshold: {threshold}')
   print()
print('WARNING: Fine-tuned models refused unsafe instructions')
print('         only ~1% of the time vs 100% for base aligned.')
```

## Ex 4: Quantization Comparison


```python
print('Quantization format comparison (32B model on H200):')
print()
for method, size, throughput, quality, target in [
    ('BF16 (baseline)','61 GB','baseline','100%','Development'),
    ('AWQ 4-bit + Marlin','~18 GB','741 tok/s (10.9×)','96%','vLLM GPU'),
    ('GPTQ 4-bit','~18 GB','712 tok/s','95%','vLLM + LoRA'),
    ('FP8 W8A8','~30 GB','33% faster','~99%','H100/H200'),
    ('GGUF Q6_K','~24 GB','CPU-friendly','98%','Ollama (quality)'),
    ('GGUF Q4_K_M','~18 GB','CPU-friendly','~94%','Ollama (speed)'),
]: print(f'  {method:22s}: {size:8s} | {throughput:18s} | {quality:5s} | {target}')
print()
print('Key: AWQ without Marlin is SLOWER than FP16!')
print('AutoAWQ archived May 2025 → use llm-compressor')
```

## Ex 5: vLLM Production Config


```python
print('vLLM production configuration:')
print()
print('vllm serve meta-llama/Llama-3.1-8B-Instruct \\')
for flag, purpose in [
    ('--enable-chunked-prefill','Optimize batch processing'),
    ('--enable-prefix-caching','Reduce TTFT for repeated system prompts'),
    ('--gpu-memory-utilization 0.9','Balance capacity vs OOM headroom'),
    ('--max-num-seqs 256','Upper limit for concurrent sequences'),
    ('--enable-lora','Enable multi-LoRA serving'),
    ('--max-loras 4','Max adapters per batch'),
]: print(f'    {flag:40s} # {purpose}')
print()
print('Performance: 793 tok/s vs Ollama 41 tok/s')
print('P99 latency: 80ms vs 673ms')
```

## Ex 6: 5 Core Production Metrics


```python
print('5 core production metrics:')
print()
for metric, target, lever in [
    ('TTFT (Time to First Token)','<200ms','Prefix caching, KV-cache routing'),
    ('TPOT/ITL (Time Per Output Token)','<100ms','Quantization, batch size'),
    ('Queue length','Near 0','Scale up GPUs, reduce max_num_seqs'),
    ('GPU memory headroom','>10% free','Reduce context, add GPUs'),
    ('Cache hit rate','>80%','Consistent hashing, sticky sessions'),
]: print(f'  {metric:35s}')
   print(f'    Target: {target:15s} | Lever: {lever}')
   print()
print('Supporting metrics: TPS, goodput, error rate, CCR')
```

## Ex 7: Deployment Strategies


```python
print('Deployment strategies for model updates:')
print()
for strategy, how, when in [
    ('Blue-green','Two environments, atomic switch, instant rollback','Low risk tolerance, need instant rollback'),
    ('Canary','Gradual: 2%→25%→75%→100%, monitor at each phase','Standard production updates'),
    ('Shadow','Mirror traffic, compare offline, no user impact','High-risk changes, new model architectures'),
]: print(f'  {strategy:12s}:')
   print(f'    How:  {how}')
   print(f'    When: {when}')
   print()
print('LLM-specific monitoring at each phase:')
print('  • Output quality scores (LLM-as-judge)')
print('  • TTFT and throughput')
print('  • Cost per request')
print('  • Hallucination / refusal rate')
```
