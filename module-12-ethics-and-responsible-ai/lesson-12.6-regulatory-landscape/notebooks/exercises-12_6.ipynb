### Step 1 · EU AI Act Risk Classifier — Where Does Your App Fall?

Automated risk tier classification based on your app’s use case.

**`01_eu_ai_act.py`** · _Block 1 of 3_

EU AI ACT RISK CLASSIFIER


```python
# EU AI ACT RISK CLASSIFIER
from dataclasses import dataclass

@dataclass
class RiskAssessment:
    tier: str; obligations: list; deadline: str; penalty: str

RISK_RULES = {
    "social_scoring": RiskAssessment("UNACCEPTABLE", ["BANNED. Cannot deploy."], "Immediate", "Up to 35M EUR or 7% revenue"),
    "biometric_id": RiskAssessment("UNACCEPTABLE", ["BANNED for real-time public use."], "Immediate", "Up to 35M EUR"),
    "hiring": RiskAssessment("HIGH", ["Risk assessment required","Human oversight mandatory","Bias testing (12.1)","Logging all decisions","Conformity assessment"], "Aug 2026", "Up to 15M EUR or 3% revenue"),
    "credit_scoring": RiskAssessment("HIGH", ["Explainability required (12.2)","Bias audit","Data governance"], "Aug 2026", "Up to 15M EUR"),
    "chatbot": RiskAssessment("LIMITED", ["Disclose AI use to users","Label AI-generated content"], "Aug 2025", "Up to 7.5M EUR"),
    "recommendation": RiskAssessment("MINIMAL", ["No mandatory obligations","Voluntary code of conduct encouraged"], "N/A", "N/A"),
    "content_gen": RiskAssessment("LIMITED", ["Label as AI-generated","Watermark if deepfake (12.5)"], "Aug 2025", "Up to 7.5M EUR"),
}

def classify(use_case):
    r = RISK_RULES.get(use_case)
    if not r: return f"Unknown use case. Manual review needed."
    return r

print("EU AI Act Risk Classifier:
")
for uc in ["chatbot","hiring","recommendation","social_scoring","content_gen"]:
    r = classify(uc)
    icon = {"UNACCEPTABLE":"❌","HIGH":"🔴","LIMITED":"🟠","MINIMAL":"🟢"}[r.tier]
    print(f"  {icon} {uc:18s} [{r.tier:14s}] Deadline: {r.deadline}")
    for o in r.obligations: print(f"      {o}")
```

> **What just happened?** Your chatbot = LIMITED risk (disclose AI use). Your hiring tool = HIGH risk (bias testing, human oversight, logging, conformity assessment). Your recommendation engine = MINIMAL (no obligations). The EU AI Act is law since Aug 2024. High-risk obligations start Aug 2026. Non-compliance: up to 35M EUR or 7% of global revenue.


### Step 2 · Google SAIF + US Policy + Industry Self-Regulation

Map the global regulatory landscape to practical compliance steps.

**`02_global_frameworks.py`** · _Block 2 of 3_

GLOBAL AI GOVERNANCE FRAMEWORKS


```python
# GLOBAL AI GOVERNANCE FRAMEWORKS

FRAMEWORKS = {
    "EU AI Act": {
        "type": "Binding law", "status": "In force (Aug 2024)",
        "key_requirements": ["Risk-tier classification", "High-risk: conformity assessment",
            "Transparency: label AI content", "GPAI models: technical documentation"],
        "your_action": "Classify your app (script 01). Implement required tier obligations.",
    },
    "Google SAIF": {
        "type": "Industry framework", "status": "Voluntary (2023)",
        "key_requirements": ["Secure AI development lifecycle", "Red-teaming and adversarial testing",
            "Supply chain security for ML models", "Monitoring deployed models"],
        "your_action": "Add red-teaming (9.11) + monitoring (10.8) + model supply chain checks.",
    },
    "US Executive Order 14110": {
        "type": "Executive guidance", "status": "Active (Oct 2023, evolving)",
        "key_requirements": ["Safety testing for dual-use models", "Watermarking AI content",
            "Reporting for large training runs", "Civil rights protections"],
        "your_action": "Implement watermarking (12.5). Document training data provenance.",
    },
    "India DPDP Act": {
        "type": "Data protection law", "status": "Enacted 2023, rules pending",
        "key_requirements": ["Consent for personal data processing", "Data localization for certain categories",
            "Right to erasure", "Breach notification within 72 hours"],
        "your_action": "PII detection (12.4). Anonymization pipeline. Data retention policy.",
    },
}

print("Global AI Governance Frameworks:
")
for name, f in FRAMEWORKS.items():
    print(f"  {name} ({f['type']}) — {f['status']}")
    for r in f["key_requirements"]: print(f"    - {r}")
    print(f"    ACTION: {f['your_action']}
")
```

### Step 3 · Compliance Mapper — From Regulation to Code

Map each regulatory requirement to the specific lesson and code that implements it.

**`03_compliance_mapper.py`** · _Block 3 of 3_

COMPLIANCE MAPPER — Regulation to code


```python
# COMPLIANCE MAPPER — Regulation to code

COMPLIANCE_MAP = [
    ("Bias testing",           "EU AI Act (High)",      "12.1 BiasTestSuite"),
    ("Explainability",         "EU AI Act (High), GDPR", "12.2 SHAP + LIME"),
    ("AI content labeling",    "EU AI Act (Limited)",   "12.5 C2PA provenance"),
    ("Deepfake watermarking",  "EU AI Act + US EO",     "12.5 TextWatermarker"),
    ("PII protection",         "GDPR + India DPDP",     "12.4 PIIDetector + Anonymizer"),
    ("Training data audit",    "EU AI Act (GPAI)",      "12.3 DataSource audit"),
    ("Monitoring deployed AI", "Google SAIF",            "10.8 AIMetrics + AlertEngine"),
    ("Red-teaming",            "Google SAIF + US EO",   "9.11 AgentGuardrails"),
    ("Incident response",      "EU AI Act (High)",      "10.9 Playbooks + PostMortem"),
    ("Human oversight",        "EU AI Act (High)",      "9.6 approval gates, 10.10 feedback"),
    ("Cost transparency",      "Internal governance",   "11.1 CostTracker + dashboard"),
]

print("Compliance Mapper: Regulation → Code
")
print(f"  {'Requirement':<25s} {'Regulation':<22s} {'Implementation'}")
print(f"  {'='*75}")
for req, reg, impl in COMPLIANCE_MAP:
    print(f"  {req:<25s} {reg:<22s} {impl}")
print(f"
  This course implements {len(COMPLIANCE_MAP)} regulatory requirements in code.")
print(f"  Every requirement maps to a specific lesson and Python class.")
```
