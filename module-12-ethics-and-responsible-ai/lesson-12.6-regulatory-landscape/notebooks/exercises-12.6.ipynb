## Ex 1: Top-Level Map (2026)


```python
print('AI regulation by jurisdiction (2026):')
for jur, primary, status in [
    ('EU',     'AI Act',                                 'in force, phased through 2026'),
    ('US',     'Exec orders + state laws + sectoral',    'patchwork, evolving'),
    ('UK',     'Pro-innovation sectoral approach',       'sector regulators leading'),
    ('China',  'GenAI Interim Measures + algo registry', 'pre-launch security assessment'),
    ('India',  'DPDP + IT Act + sectoral + DIA draft',   'consolidation pending'),
    ('Brazil', 'Marco Civil + AI bill PL 2338',          'in committee'),
    ('Canada', 'AIDA (proposed)',                        'C-27 progressing'),
    ('Japan',  'METI guidelines + voluntary',            'soft-law approach'),
]:
    print(f'  {jur:8s}: {primary:42s} | {status}')
```

## Ex 2: EU AI Act Risk Tiers


```python
print('EU AI Act risk tiers:')
for tier, examples, obligations in [
    ('Unacceptable',  'social scoring, real-time biometric ID', 'BANNED'),
    ('High',          'employment, credit, education, law enf', 'risk mgmt + data + oversight + monitoring'),
    ('Limited',       'chatbots, deepfakes',                    'transparency: disclose AI'),
    ('Minimal',       'spam filters, AI in games',              'no specific obligation'),
    ('GPAI (general)','foundation models',                       'transparency + safety + copyright'),
]:
    print(f'  {tier:14s}: {examples:42s} | {obligations}')
```

## Ex 3: India Regulatory Stack


```python
print('India AI regulation stack (2026):')
for instrument, scope, status in [
    ('DPDP Act 2023',                'personal data',                        'in force; rules being notified'),
    ('IT Act 2000 + Intermediary R 2021','platforms + takedown SLAs',         'in force'),
    ('RBI sectoral (banking/fintech)',  'data residency, model risk',         'in force'),
    ('SEBI sectoral (capital markets)', 'algo trading, investor protection',  'in force'),
    ('IRDAI sectoral (insurance)',      'AI underwriting, claims',            'in force'),
    ('MeitY guidelines',                'broad responsible AI',                'advisory'),
    ('ECI campaign-AI guidelines',      'election content disclosure',         '2024'),
    ('NITI Aayog principles',           'responsible AI baseline',             'soft-law'),
    ('Digital India Act (DIA) draft',   'consolidate IT Act + AI duties',      '2024-26 expected'),
    ('BIS adoption ISO/IEC 42001',      'AI management system standard',       'in progress'),
]:
    print(f'  {instrument:36s}: {scope:36s} | {status}')
```

## Ex 4: NIST AI RMF Functions


```python
print('NIST AI RMF four functions:')
for fn, what, examples in [
    ('GOVERN',  'org policies + roles + accountability', 'AI policy doc, ethics committee, model owners'),
    ('MAP',     'context + risks + stakeholders',         'feature DPIA, threat model, user impact'),
    ('MEASURE', 'eval + monitoring',                       'bench, fairness eval, eval-on-prod, drift'),
    ('MANAGE',  'mitigations + incident response',         'guardrails, kill switch, runbooks, postmortem'),
]:
    print(f'  {fn:10s}: {what:36s} | examples: {examples}')
print()
print('NIST AI 600-1: GenAI profile addendum (2024). Use for procurement + structured baseline.')
```

## Ex 5: ISO/IEC 42001 — AI Management System


```python
print('ISO/IEC 42001 (AIMS) clauses overview:')
for clause, what in [
    ('4. Context',          'org context + AI scope'),
    ('5. Leadership',       'policy + roles + accountability'),
    ('6. Planning',         'risk + opportunity + objectives'),
    ('7. Support',          'resources + competence + comms'),
    ('8. Operation',        'lifecycle controls + supplier mgmt'),
    ('9. Performance eval', 'monitoring + audit + management review'),
    ('10. Improvement',     'nonconformity + corrective action'),
]:
    print(f'  {clause:24s}: {what}')
print()
print('Cert lead time 6-12 months. Procurement requirement for enterprise + govt deals.')
```

## Ex 6: US State Patchwork


```python
print('US state AI laws (selected, 2026):')
for state, law, scope in [
    ('Colorado',  'CO AI Act (Feb 2026)',     'high-risk consequential decisions'),
    ('NY',        'NYC Local Law 144',         'AI hiring bias audit'),
    ('California','SB 1047 vetoed; successor likely', 'frontier model safety'),
    ('Illinois',  'BIPA',                       'biometrics + consent'),
    ('Texas',     'TX AI law + employment',    'workplace AI'),
    ('Utah',      'AI Policy Act',             'disclosure + consumer protection'),
]:
    print(f'  {state:10s} | {law:30s} | {scope}')
```

## Ex 7: Compliance Matrix Operationalization


```python
print('Compliance matrix — minimum columns:')
for col, why in [
    ('Jurisdiction',       'EU / US-CO / India-RBI / etc.'),
    ('Obligation',         'specific clause or article'),
    ('Applies to feature', 'which product features in scope'),
    ('Implementation',     'description + link to control'),
    ('Owner',              'team + RACI'),
    ('Status',             'planned / in-progress / done / verified'),
    ('Last reviewed',      'date + reviewer'),
    ('Evidence',           'link to artifacts (DPIA, model card, audit log)'),
]:
    print(f'  {col:24s} -> {why}')
print()
print('Quarterly review with legal. Highest-bar wins on conflicting rules.')
```
