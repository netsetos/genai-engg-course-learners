### Step 1 · AI Policy Template — Your Organization’s AI Rules

A structured policy document every AI team needs. Fill in the blanks.

**`01_ai_policy.py`** · _Block 1 of 3_

AI POLICY TEMPLATE — Organizational AI governance


```python
# AI POLICY TEMPLATE — Organizational AI governance
from dataclasses import dataclass, field
from datetime import datetime

@dataclass
class AIPolicy:
    org_name: str
    version: str
    effective_date: str
    approved_models: list[str]
    prohibited_uses: list[str]
    data_rules: list[str]
    review_frequency: str
    incident_contact: str

    def render(self):
        print(f"{'='*60}")
        print(f"AI USAGE POLICY — {self.org_name}")
        print(f"Version {self.version} | Effective: {self.effective_date}")
        print(f"{'='*60}")
        print(f"
Approved Models:")
        for m in self.approved_models: print(f"  - {m}")
        print(f"
Prohibited Uses:")
        for p in self.prohibited_uses: print(f"  X {p}")
        print(f"
Data Rules:")
        for d in self.data_rules: print(f"  - {d}")
        print(f"
Review: {self.review_frequency} | Incidents: {self.incident_contact}")

policy = AIPolicy(
    org_name="Netsetos", version="2.0", effective_date="2026-04-01",
    approved_models=["Gemini 2.0 Flash", "Gemini 2.5 Pro", "Llama 3.1 (self-hosted)"],
    prohibited_uses=["Automated hiring decisions without human review",
        "Processing student PII through third-party LLMs",
        "Generating content impersonating real individuals",
        "Using AI outputs as sole basis for disciplinary action"],
    data_rules=["No PII in prompts to cloud LLMs (use anonymization pipeline 12.4)",
        "All LLM calls logged with cost tracking (11.1)",
        "Student data processed only on approved models",
        "Training data audited quarterly (12.3)"],
    review_frequency="Quarterly", incident_contact="[email protected]")
policy.render()
```

> **What just happened?** A complete AI policy in a Python dataclass: approved models, prohibited uses, data handling rules, review schedule. Every organization using AI needs this document. It sets boundaries BEFORE something goes wrong. The policy references specific course tools: anonymization (12.4), cost tracking (11.1), data audit (12.3).


### Step 2 · Model Card Generator — Document Every Deployed Model

Standardized documentation: what the model does, its limits, and known biases.

**`02_model_card.py`** · _Block 2 of 3_

MODEL CARD GENERATOR — Document every deployed model


```python
# MODEL CARD GENERATOR — Document every deployed model
from dataclasses import dataclass, field
import json

@dataclass
class ModelCard:
    name: str
    version: str
    provider: str
    use_case: str
    intended_users: str
    input_type: str
    output_type: str
    limitations: list[str]
    known_biases: list[str]
    eval_metrics: dict
    ethical_considerations: list[str]
    data_sources: list[str]
    last_evaluated: str

    def to_json(self): return json.dumps(self.__dict__, indent=2)
    def render(self):
        print(f"MODEL CARD: {self.name} v{self.version}")
        print(f"Provider: {self.provider} | Use: {self.use_case}")
        print(f"Users: {self.intended_users}")
        print(f"Eval: {self.eval_metrics}")
        print(f"Limitations: {', '.join(self.limitations)}")
        print(f"Known biases: {', '.join(self.known_biases)}")

card = ModelCard(
    name="Netsetos Support Bot", version="3.1", provider="Gemini 2.0 Flash",
    use_case="Answer student questions about courses, pricing, refunds",
    intended_users="Prospective and enrolled Netsetos students",
    input_type="Natural language text (English, Hindi)",
    output_type="Natural language text with source citations",
    limitations=["Cannot process images", "May hallucinate pricing if not in RAG context",
                 "English responses higher quality than Hindi"],
    known_biases=["Slightly longer responses for English queries",
                  "Tested for name bias: no significant difference (p=0.42)"],
    eval_metrics={"accuracy":"92%", "hallucination_rate":"3%", "avg_latency":"1.2s"},
    ethical_considerations=["Not used for admission decisions", "Human escalation available"],
    data_sources=["Netsetos course catalog", "FAQ database", "Pricing API"],
    last_evaluated="2026-03-15")
card.render()
```

### Step 3 · Risk Assessment + Audit Framework

Structured risk assessment before deployment. Automated audit checks post-deployment.

**`03_risk_audit.py`** · _Block 3 of 3_

RISK ASSESSMENT + AUDIT FRAMEWORK


```python
# RISK ASSESSMENT + AUDIT FRAMEWORK
from dataclasses import dataclass

@dataclass
class RiskItem:
    category: str; risk: str; likelihood: str; impact: str; mitigation: str; status: str

class AIAudit:
    def __init__(self):
        self.checks = [
            ("Bias",         ["Bias test suite run (12.1)", "No p<0.05 bias found", "Tested monthly"]),
            ("Privacy",      ["PII detector active (12.4)", "Anonymization pipeline deployed", "No PII in LLM logs"]),
            ("Security",     ["Input guardrails active (9.11)", "Output filtering enabled", "Rate limiting configured"]),
            ("Transparency", ["AI use disclosed to users", "Model card published", "Confidence indicators shown (10.10)"]),
            ("Reliability",  ["Circuit breaker configured (10.2)", "Fallback providers set", "Monitoring active (10.8)"]),
            ("Cost",         ["Cost tracking enabled (11.1)", "Per-team budgets set", "Anomaly alerts configured"]),
            ("Legal",        ["Training data audited (12.3)", "EU AI Act tier classified (12.6)", "Terms of service updated"]),
        ]

    def run(self, implemented):
        print("AI GOVERNANCE AUDIT
")
        total, passed = 0, 0
        for cat, items in self.checks:
            print(f"  {cat}:")
            done = implemented.get(cat, [])
            for item in items:
                total += 1; ok = item in done
                if ok: passed += 1
                print(f"    {'✓' if ok else '✗'} {item}")
        score = passed/total*100
        print(f"
  Score: {passed}/{total} ({score:.0f}%)")
        print(f"  {'PASS: ready for production' if score>=80 else 'FAIL: address gaps before launch'}")

audit = AIAudit()
audit.run({
    "Bias": ["Bias test suite run (12.1)", "No p<0.05 bias found"],
    "Privacy": ["PII detector active (12.4)", "Anonymization pipeline deployed", "No PII in LLM logs"],
    "Security": ["Input guardrails active (9.11)", "Output filtering enabled"],
    "Transparency": ["AI use disclosed to users", "Model card published"],
    "Reliability": ["Circuit breaker configured (10.2)", "Fallback providers set", "Monitoring active (10.8)"],
    "Cost": ["Cost tracking enabled (11.1)", "Per-team budgets set"],
    "Legal": ["Training data audited (12.3)"],
})
```
