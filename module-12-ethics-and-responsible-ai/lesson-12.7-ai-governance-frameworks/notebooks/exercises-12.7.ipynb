## Ex 1: Governance Roles


```python
print('Required / recommended roles:')
for role, mandate, when in [
    ('Data Protection Officer',   'DPDP / GDPR mandated', 'any org with personal data'),
    ('Chief AI Officer',          'strategy + accountability', 'mid-to-large orgs'),
    ('AI Ethics Committee',       'review high-risk features', '5-9 cross-functional'),
    ('Model Owner',                'per deployed model',        'always'),
    ('Risk Owner',                 'per identified risk',       'always'),
    ('Responsible AI Team',        'tooling + training',        'mature orgs'),
    ('Incident Commander',         'per SEV1/2 incident',        'always (10.9)'),
]:
    print(f'  {role:26s}: {mandate:30s} | when: {when}')
```

## Ex 2: AI Policy Doc Skeleton


```python
print('AI policy doc sections:')
for n, section in enumerate([
    'Scope: which AI systems covered',
    'Principles: accountability, fairness, transparency, privacy, safety',
    'Approved models + prohibited uses',
    'Data handling: residency, PII, consent, retention',
    'Development lifecycle: DPIA -> model card -> eval -> deploy',
    'Monitoring + incident response',
    'Roles: DPO, CAIO, model owners, committee',
    'Review cadence: quarterly; signed by CEO',
    'Links to sub-policies (DPIA template, incident runbook, etc.)',
], 1):
    print(f'  {n}. {section}')
print()
print('5-15 pages; in Git; versioned; reviewed quarterly.')
```

## Ex 3: Model Card Template


```python
print('Model card sections (Mitchell et al. 2019 + HF standard):')
for section, what in [
    ('Model Details',       'name, version, type, architecture'),
    ('Intended Use',         'primary use + users + environment'),
    ('Training Data',        'source, size, preprocessing, licenses'),
    ('Evaluation',           'metrics + datasets + results'),
    ('Quantitative Analyses','per-group performance (fairness)'),
    ('Ethical Considerations','risks, harms, misuses'),
    ('Caveats & Limitations','known failure modes'),
    ('Environmental Impact', 'compute hours, CO2e'),
]:
    print(f'  {section:24s}: {what}')
```

## Ex 4: Risk Assessment Method (ISO 31000 / NIST)


```python
print('Risk assessment 4 steps:')
for step, what, output in [
    ('1. Identify',  'what can go wrong (bias, privacy, IP, safety, misuse)', 'risk register'),
    ('2. Analyze',   'likelihood \u00d7 impact',                                     'heatmap per risk'),
    ('3. Evaluate',  'above tolerance? acceptable? residual?',                 'accept / treat list'),
    ('4. Treat',     'avoid / reduce / transfer / accept',                     'mitigations + owners'),
]:
    print(f'  {step:14s}: {what:54s} | output: {output}')
```

## Ex 5: Ethics Committee Cadence


```python
print('AI ethics committee setup:')
for aspect, detail in [
    ('Size',                  '5-9 members'),
    ('Composition',            'Legal, Product, Engineering/ML, Security, HR, ext. ethicist'),
    ('Cadence',                'monthly early-stage; quarterly when stable'),
    ('Inputs',                  'new high-risk features, incidents, policy updates, audit findings'),
    ('Outputs',                 'approvals, conditions, action items, dissent recorded'),
    ('Independence',            'separate from product P&L for credibility'),
    ('Decision log',            'immutable, evidentiary'),
]:
    print(f'  {aspect:16s}: {detail}')
```

## Ex 6: AI-Specific Incident Types


```python
print('AI-specific incident categories (extend 10.9 severity):')
for cat, example, primary_owner in [
    ('Fairness',    'disparate outcome by group',          'Ethics + ML'),
    ('Privacy',      'PII memorization or leak',            'DPO + Security'),
    ('IP',           'copyright match, license violation',  'Legal + ML'),
    ('Safety',       'harmful output, dangerous advice',    'Product + ML'),
    ('Security',     'prompt injection, model extraction',  'Security'),
    ('Compliance',   'regulation violation surfaced',        'Legal + DPO'),
]:
    print(f'  {cat:14s}: {example:36s} | owner: {primary_owner}')
```

## Ex 7: Week-1 Starter Kit Checklist


```python
print('Week-1 governance starter kit (8 artifacts):')
for n, item in enumerate([
    'AI policy v0.1 (even 1 page is OK)',
    'Model card template in repo (HF-style)',
    'DPIA template + process',
    'Model owners assigned per deployed model',
    'AI ethics committee of 4 (can grow later)',
    'Incident runbooks for fairness + privacy + IP + safety',
    'Compliance matrix (12.6) started',
    'Quarterly governance review on calendar',
], 1):
    print(f'  {n}. {item}')
print()
print('Ship Week 1; evolve monthly. Principles without an operating model are vibes.')
```
