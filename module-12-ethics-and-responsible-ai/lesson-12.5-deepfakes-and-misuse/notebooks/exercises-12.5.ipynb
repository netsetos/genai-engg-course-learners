## Ex 1: Threat Modalities + Tools


```python
print('Deepfake threat modalities (2026):')
for modality, tools, typical_attack in [
    ('Voice',  'ElevenLabs, XTTS, Resemble',         'CEO vishing, KYC voice bypass'),
    ('Video',  'DeepFaceLab, Real-time face swap',   'executive impersonation calls'),
    ('Image',  'Midjourney, FLUX, SDXL',              'fabricated evidence, NSFW'),
    ('Text',   'GPT-5, Sonnet, local Llama',          'phishing at scale, bot swarms'),
    ('Mixed',  'lip-sync + face swap + voice',       'fake video calls'),
]:
    print(f'  {modality:8s}: {tools:34s} | attack: {typical_attack}')
```

## Ex 2: C2PA Provenance Stack


```python
print('C2PA content provenance stack:')
for layer, what, status in [
    ('Capture manifest', 'camera/mic signs origin + timestamp', 'shipping in iPhone, Nikon'),
    ('AI-gen manifest',   'generator signs with model ID + ts',  'OpenAI, Microsoft, Adobe adopting'),
    ('Edit manifest',     'each edit appends signed entry',      'Adobe Photoshop, Premiere'),
    ('Verify on display',  'browser/platform shows badge',        'Microsoft Edge, Content Auth'),
    ('Revocation',         'key rotation + CRLs',                 'still maturing'),
]:
    print(f'  {layer:22s}: {what:42s} | {status}')
```

## Ex 3: Watermarking Techniques


```python
print('Watermarking by modality:')
for modality, technique, robustness in [
    ('Image',  'SynthID (Google), Stable Signature',         'survives compression, not crop'),
    ('Audio',  'frequency-domain spread spectrum',            'survives re-encoding, not paraphrase'),
    ('Text',   'biased token sampling',                       'fragile; lost on paraphrase'),
    ('Video',  'per-frame + audio watermark stack',            'moderate'),
]:
    print(f'  {modality:8s}: {technique:42s} | robustness: {robustness}')
print()
print('Treat as deterrent + post-hoc detection. Provenance (C2PA) remains stronger.')
```

## Ex 4: KYC Liveness Defenses


```python
print('KYC liveness layers (defense in depth):')
for layer, how, bypass_cost in [
    ('Active challenge-response','blink, head turn, read random digit','real-time video clone needed'),
    ('3D depth check',           'TrueDepth, IR structured light',      'physical mask or display hack'),
    ('rPPG blood flow',          'skin micro-color change detection',    'hard to fake in real-time'),
    ('Anti-replay',              'sensor fingerprint + device binding', 'requires device compromise'),
    ('Session randomness',       'never repeat exact challenge',         'breaks pre-recorded spoofs'),
]:
    print(f'  {layer:30s}: {how:40s} | bypass: {bypass_cost}')
```

## Ex 5: Refusal Taxonomy for Your LLM Product


```python
print('Hard-refusal categories your LLM should handle:')
for category, example_prompt, response in [
    ('Phishing drafts',           '"write urgent email as CEO to wire $50k"', 'refuse + explain'),
    ('Real-person impersonation', '"write as Narendra Modi announcing..."',   'refuse; offer generic alt'),
    ('NSFW of real minors',        '(any such prompt)',                         'refuse; log + escalate'),
    ('Weapon / bio uplift',        '"synthesize nerve agent"',                   'refuse'),
    ('Sensitive PII lookup',       '"what is X person\'s Aadhaar"',              'refuse'),
    ('Election manipulation',      '"fake quote from a politician"',             'refuse + label AI'),
]:
    print(f'  {category:28s}: {example_prompt:42s} | {response}')
```

## Ex 6: Indian Legal Landscape


```python
print('Indian laws touching deepfakes (2026):')
for law, applies_to in [
    ('IT Act 66E',                'voyeurism / violation of privacy'),
    ('IT Act 67',                 'obscene material in electronic form'),
    ('IPC 419',                   'impersonation'),
    ('IPC 420',                   'cheating'),
    ('IPC 500',                   'defamation'),
    ('Intermediary Guidelines 2021','platform takedown SLA 24-72h'),
    ('ECI campaign AI guidelines', 'party/candidate disclosure of AI content'),
    ('DPDP Act 2023',              'personal data misuse in training / generation'),
    ('Pending deepfake law',       'discussed in Parliament 2024-26'),
]:
    print(f'  {law:32s} -> {applies_to}')
```

## Ex 7: Week-1 Product Responsibility Checklist


```python
print('Week-1 misuse-mitigation checklist (every GenAI product):')
for n, item in enumerate([
    'Publish Responsible AI / Acceptable Use policy',
    'Refusal training + output filters on known misuse categories',
    'C2PA / SynthID signing on all generated outputs',
    'Abuse-report channel (form + email) + 72h SLA',
    'Rate limits on suspicious patterns (impersonation prompts, volume)',
    'Monthly review: top refusals, emerging misuse, adversarial jailbreaks',
], 1):
    print(f'  {n}. {item}')
```
