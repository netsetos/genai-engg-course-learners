### Step 1 · Training Data Legality — What Can You Train On?

The legal landscape: public domain, fair use, licensed, and the gray areas.

**`01_training_data_audit.py`** · _Block 1 of 3_

TRAINING DATA AUDIT — Classify data sources by legal risk


```python
# TRAINING DATA AUDIT — Classify data sources by legal risk
from dataclasses import dataclass
from enum import Enum

class Risk(Enum):
    LOW = "low"; MEDIUM = "medium"; HIGH = "high"

@dataclass
class DataSource:
    name: str; license: str; risk: Risk; notes: str

SOURCES = [
    DataSource("Wikipedia", "CC BY-SA 3.0", Risk.LOW, "Freely usable with attribution"),
    DataSource("arXiv papers", "Varies", Risk.LOW, "Most are open access"),
    DataSource("GitHub public repos", "Varies (MIT/GPL/none)", Risk.MEDIUM, "Check each repo license"),
    DataSource("Reddit posts", "TOS-dependent", Risk.MEDIUM, "Reddit API TOS restricts use"),
    DataSource("News articles", "Copyrighted", Risk.HIGH, "NYT v OpenAI lawsuit precedent"),
    DataSource("Books (pirated)", "Copyrighted", Risk.HIGH, "Clear infringement. Never use."),
    DataSource("Common Crawl", "Mixed", Risk.MEDIUM, "Contains copyrighted content"),
    DataSource("Licensed datasets", "Commercial", Risk.LOW, "Paid license = clear rights"),
]

print("Training Data Risk Audit:\n")
for s in SOURCES:
    icon = {"low":"✅","medium":"⚠️","high":"❌"}[s.risk.value]
    print(f"  {icon} {s.name:25s} [{s.risk.value:6s}] {s.license:20s} {s.notes}")
```

### Step 2 · Generated Content Ownership — Who Owns AI Output?

Current legal status across jurisdictions. Practical guidelines.

**`02_ownership_checker.py`** · _Block 2 of 3_

GENERATED CONTENT OWNERSHIP CHECKER


```python
# GENERATED CONTENT OWNERSHIP CHECKER

OWNERSHIP_RULES = {
    "US": {"status":"No copyright for purely AI-generated content",
           "human_element":"Copyright possible if human provides substantial creative input",
           "case_law":"Thaler v Perlmutter (2023): AI cannot be inventor/author"},
    "EU": {"status":"Varies by member state. Generally requires human author.",
           "ai_act":"EU AI Act requires transparency: label AI-generated content"},
    "India": {"status":"Copyright Act requires human author. AI output unclear.",
             "practical":"Treat AI output as tool-assisted human creation"},
}

def check_ownership(content_type, human_involvement, jurisdiction):
    rules = OWNERSHIP_RULES.get(jurisdiction, {})
    if human_involvement == "high":
        return "Likely copyrightable: substantial human creative input"
    elif human_involvement == "medium":
        return "Gray area: document human creative choices thoroughly"
    return "Likely NOT copyrightable: purely AI-generated"

print("Content Ownership:\n")
for case in [("Blog post","high","US"), ("AI image","low","US"), ("Code","medium","India")]:
    print(f"  {case[0]:15s} (human: {case[1]:6s}, {case[2]}): {check_ownership(*case)}")
```

### Step 3 · Compliance Checklist — Before You Ship

Practical checklist for responsible AI deployment.

**`03_compliance_checklist.py`** · _Block 3 of 3_

COMPLIANCE CHECKLIST FOR AI DEPLOYMENT


```python
# COMPLIANCE CHECKLIST FOR AI DEPLOYMENT

CHECKLIST = [
    ("Training Data", [
        "All data sources documented with licenses",
        "No pirated books, scraped copyrighted articles",
        "Opted-out data removed (robots.txt compliance)",
        "PII removed or anonymized from training data",]),
    ("Generated Content", [
        "AI-generated content labeled as such",
        "No attribution of fake quotes to real people",
        "Output filtering for copyrighted reproductions",
        "User retains rights to their prompts",]),
    ("Disclosure", [
        "Users know they are interacting with AI",
        "Model name and version disclosed",
        "Limitations and known biases documented",
        "Data retention policy published",]),
    ("Legal", [
        "Terms of service cover AI-generated content",
        "Indemnification clause for AI errors",
        "Compliance with local AI regulations (EU AI Act)",]),
]

print("AI Deployment Compliance Checklist:\n")
total, done = 0, 0
for category, items in CHECKLIST:
    print(f"  {category}:")
    for item in items:
        total += 1; print(f"    [ ] {item}")
print(f"\n  Total: {total} items. Complete ALL before production launch.")
```
