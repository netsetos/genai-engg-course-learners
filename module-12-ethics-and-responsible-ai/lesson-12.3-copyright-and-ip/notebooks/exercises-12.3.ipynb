## Ex 1: Active Lawsuit Landscape (2026)


```python
print('Active or recently-decided AI copyright cases:')
for case, jurisdiction, status, takeaway in [
    ('NYT v OpenAI/Microsoft',  'US (SDNY)',  'pending',     'verbatim reproduction central'),
    ('Authors Guild v OpenAI',  'US (SDNY)',  'pending',     'class action by published authors'),
    ('Getty v Stability',       'US + UK',    'pending',     'image gen + watermark presence'),
    ('Doe v GitHub Copilot',    'US',         'mostly dismissed','licensed code training'),
    ('EU AI Act',               'EU',         'in force 2024-26','TDM opt-out + transparency'),
    ('Indian Copyright Amend',  'India',      'discussion',  'TDM exception unclear'),
]:
    print(f'  {case:28s} | {jurisdiction:10s} | {status:18s} | {takeaway}')
```

## Ex 2: Output Ownership by Jurisdiction


```python
print('Pure AI-generated output ownership (2026):')
for jurisdiction, status, exception in [
    ('US (Copyright Office)', 'NO copyright on pure AI output', 'human-authored elements protected'),
    ('UK',                    'NO copyright on pure AI output', 'similar to US'),
    ('India (Copyright Office 2024)','similar to US',           'awaiting amendment for clarity'),
    ('China',                 'YES with human contribution',    'broader than West'),
    ('EU',                    'NO; needs human author',         'AI as tool only'),
]:
    print(f'  {jurisdiction:30s}: {status:32s} | exception: {exception}')
print()
print('Implication: document human contribution (prompt design, edit, curation) to claim partial copyright.')
```

## Ex 3: Memorization Detection


```python
print('Memorization detection methods:')
for method, what, false_positive_risk in [
    ('Exact substring match (50+ tokens)','direct verbatim',         'low'),
    ('N-gram overlap > 30%',              'paraphrased reproduction','medium'),
    ('Embedding similarity > 0.95',        'semantic reproduction',   'higher (catches valid summarization)'),
    ('Known-passage prompt probe',         'targeted memorization',   'low; needs canary set'),
    ('Watermark detection (if present)',   'image/audio outputs',     'depends on model'),
]:
    print(f'  {method:36s}: {what:28s} | FP risk: {false_positive_risk}')
```

## Ex 4: Provider Copyright Shields (2026)


```python
print('Provider copyright shield comparison (READ FINE PRINT):')
for prov, coverage, conditions in [
    ('OpenAI',     'defense costs + judgments', 'safety features ON, no intentional infringement'),
    ('Anthropic',  'defense costs',             'enterprise tier, defaults on'),
    ('Microsoft',  'defense + indemnity',       'Copilot Copyright Commitment, guardrails on'),
    ('Google',     'defense + indemnity',       'Vertex AI, generated outputs only (not inputs)'),
    ('AWS Bedrock','defense + indemnity',       'capped, specific models, guardrails on'),
]:
    print(f'  {prov:14s}: {coverage:30s} | conditions: {conditions}')
print()
print('Always: caps, exclusions for fine-tunes, requires safety defaults on. Verify per contract.')
```

## Ex 5: RAG Copyright Risk


```python
print('Two RAG copyright surfaces:')
for surface, issue, mitigation in [
    ('Storage in vector DB','reproduction = restricted right',  'license docs, customer-owned only'),
    ('Generated quotes',    'derivative work',                  'cite + limit verbatim &lt; 50 words'),
    ('Cross-tenant leakage','one tenant sees another\'s docs',  'strict tenant isolation'),
    ('Public web scrape',   'unlicensed corpus',                'use Common Crawl + opt-out compliance'),
    ('Take-down requests',  'DMCA-style process needed',         'documented + actionable within 72h'),
]:
    print(f'  {surface:22s}: {issue:36s} | mitigate: {mitigation}')
```

## Ex 6: License Compatibility for Code Training


```python
print('Code training compatibility (2026 community consensus):')
for license, training_ok, output_attr in [
    ('MIT',             'YES',  'optional but courteous'),
    ('Apache 2.0',      'YES',  'optional'),
    ('BSD',             'YES',  'optional'),
    ('GPLv2/v3',        'CONTESTED', 'risk of viral output'),
    ('LGPL',            'CONTESTED', 'similar GPL concerns'),
    ('AGPL',            'AVOID',     'most viral'),
    ('Proprietary',     'NO',        'cannot train'),
]:
    print(f'  {license:20s}: train OK = {training_ok:10s} | output: {output_attr}')
```

## Ex 7: 8-Point Risk Management Checklist


```python
print('8-point IP risk management checklist:')
for n, item in enumerate([
    'Document training + fine-tune data provenance per dataset',
    'Enable provider copyright shields (defaults ON, contract verified)',
    'Output verbatim-match detector (50+ tokens) against known corpora',
    'RAG citations on every claim; limit quote &lt; 50 words',
    'License all third-party content you ingest (CC, CC-BY, public domain)',
    'Audit log every generation with provenance + filters applied',
    'DMCA-style takedown process documented + 72h SLA',
    'Customer indemnity in T&Cs where contractually viable',
], 1):
    print(f'  {n}. {item}')
```

---
## Recap
Training on copyrighted data: high legal risk; multiple major lawsuits pending.
Output: NO copyright on pure AI; document human contribution.
Provider shields cover defense IF defaults on; read fine print.
RAG: license + cite + limit quote length. Code: MIT/Apache OK, GPL contested.
Fine-tuning: ONLY use customer-licensed / public-domain / explicitly-licensed.
8-point checklist: provenance, shields, filter, citations, license, audit, takedown, indemnity.
