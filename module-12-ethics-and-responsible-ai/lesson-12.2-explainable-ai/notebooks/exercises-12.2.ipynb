## Ex 1: 3 Layers of Explanation


```python
print('Three observable layers for production LLM explainability:')
for layer, what, tool in [
    ('Inputs',     'context, RAG chunks, tools, system prompt', 'OTel spans + Langfuse traces'),
    ('Reasoning',  'chain-of-thought, extended thinking',        'CoT collapsible UI (10.10)'),
    ('Citations',  'source spans grounding each claim',          'Anthropic citations, custom regex'),
]:
    print(f'  {layer:14s}: {what:42s} | tool: {tool}')
print()
print('Mechanistic interpretability (sparse autoencoders) is research-stage; not yet production-actionable.')
```

## Ex 2: Citations vs Explanations


```python
print('Citations vs explanations — different jobs:')
for thing, what, who_needs in [
    ('Citation',    'pointer to source backing a claim',  'all users'),
    ('Explanation', 'description of how model decided',   'users + auditors'),
    ('Trace',       'spans of internal call chain',        'developers'),
    ('Counterfactual','what would change the answer',     'debug + bias audit'),
    ('Model card',   'capability + risk + training summary','regulators + procurement'),
]:
    print(f'  {thing:14s}: {what:42s} | for: {who_needs}')
```

## Ex 3: Anthropic Citations Demo Shape


```python
print('Anthropic citations API output shape (illustrative):')
example = {
    'output_sentence': 'Loan approval requires income verification.',
    'citations': [
        {'document_index': 0, 'cited_text': 'Income verification is mandatory...', 'start_char_index': 412, 'end_char_index': 510},
        {'document_index': 1, 'cited_text': 'KYC must include income proof.',     'start_char_index': 1208, 'end_char_index': 1240},
    ]
}
import json
print(json.dumps(example, indent=2))
print()
print('UI: inline [1][2] badges with hover, click jumps to source panel highlight.')
```

## Ex 4: CoT Trustworthiness Caveat


```python
print('Why CoT is not the answer:')
for issue, example in [
    ('Confidently wrong',        'CoT correctly arithmeticizes, then asserts wrong final number'),
    ('Not always faithful',      'CoT may post-hoc rationalize; not the actual decision path'),
    ('Length != correctness',     '5-step CoT can be wrong; 1-line answer can be right'),
    ('Can leak system prompt',    'unfiltered CoT may expose internal rules / safety thresholds'),
]:
    print(f'  {issue:24s}: {example}')
print()
print('Eval on FINAL answer; use CoT as supporting context, not as the verdict.')
```

## Ex 5: Counterfactual Sensitivity Probe


```python
print('Counterfactual probes — what flips the answer?')
for swap, expected_effect in [
    ('Entity name (Priya -> Mohammad)', 'should NOT change; if it does -> bias'),
    ('Drop top RAG chunk',              'reveals which chunk dominated'),
    ('Reorder chunks',                  'should NOT change; sensitivity = position bias'),
    ('Truncate context to 50%',          'shows whether long context actually used'),
    ('Replace tool result with None',    'reveals which tool drove the conclusion'),
]:
    print(f'  swap: {swap:36s} | expected: {expected_effect}')
```

## Ex 6: Eval Faithfulness, Plausibility, Coverage


```python
print('Three eval axes for explanations:')
for axis, definition, test in [
    ('Faithfulness',  'reflects actual model process', 'edit explanation; answer should not change'),
    ('Plausibility',  'makes sense to a human',         'user study or LLM-as-judge rubric'),
    ('Coverage',      'every important claim cited',    'regex citation check; flag uncited'),
]:
    print(f'  {axis:14s}: {definition:34s} | test: {test}')
```

## Ex 7: Stakeholder Explanation Matrix


```python
print('Different stakeholders need different explanations:')
for stake, format, depth in [
    ('End user',         '1-line + sources + human-review button',  'minimal, plain language'),
    ('Customer support', 'citations + CoT + decision context',      'medium'),
    ('Developer',         'OTel trace + spans + attributions',      'deep'),
    ('Auditor / regulator','model card + training summary + logs',  'documentation-grade'),
    ('Legal',            'logged decisions + grievance pathway',     'evidentiary'),
]:
    print(f'  {stake:20s}: {format:40s} | depth: {depth}')
```
