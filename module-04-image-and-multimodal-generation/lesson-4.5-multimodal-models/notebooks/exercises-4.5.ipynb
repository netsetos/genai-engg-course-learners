## Ex 6: Multimodal RAG Concept


```python
print('Traditional OCR Pipeline:')
print('  scan → OCR → chunk → embed → retrieve → answer')
print('  ❌ Loses layout, tables, charts')
print('  ❌ OCR errors compound')
print('  ❌ Struggles with regional scripts')
print()
print('Multimodal RAG Pipeline:')
print('  scan as image → embed with ColPali → FAISS index')
print('  query → retrieve page images → VLM reads directly')
print('  ✅ Preserves layout perfectly')
print('  ✅ Handles any script/language')
print('  ✅ No OCR errors')
print()
print('ColPali: uses PaliGemma vision embeddings for retrieval')
print('ColQwen2: uses Qwen2-VL vision embeddings')
```

## Ex 7: India Cost Calculator


```python
scenarios = [
    ('KYC verification', 50000, 'Self-host 7B (DPDP compliance)'),
    ('Crop disease ID', 10000, 'Gemini Flash API'),
    ('Prescription reading', 20000, 'Self-host 7B (medical privacy)'),
    ('Product descriptions', 100000, 'Gemini Flash API'),
    ('Document digitization', 500000, 'Self-host 72B (quality)'),
]
print(f'{"Use Case":25s} {"Volume/mo":>10s} {"Best Option":>30s} {"Cost/mo":>10s}')
print('-'*80)
for name, vol, opt in scenarios:
    if 'Self-host 7B' in opt: cost = vol * 0.008 * 84  # ₹
    elif 'Gemini Flash' in opt: cost = vol * 0.0004 * 84  # ₹
    elif '72B' in opt: cost = vol * 0.002 * 84
    else: cost = vol * 0.005 * 84
    print(f'{name:25s} {vol:>10,} {opt:>30s} ₹{cost:>9,.0f}')
```
