## Ex 4: Voice Agent Pipeline


```python
print('Cascaded Voice Agent Pipeline:')
print()
stages = [
    ('Microphone', '0ms', 'Capture audio via WebRTC'),
    ('VAD (Silero)', '10ms', 'Detect speech vs silence'),
    ('STT (Deepgram)', '200ms', 'Transcribe speech to text'),
    ('Turn Detection', '50ms', 'Detect end of user turn'),
    ('LLM (GPT-4o)', '500ms', 'Generate response'),
    ('TTS (ElevenLabs)', '150ms', 'Synthesize speech'),
    ('Network + Buffer', '100ms', 'Transport to client'),
]
total = 0
for stage, latency, desc in stages:
    ms = int(latency.replace('ms','')) if latency != '0ms' else 0
    total += ms
    print(f'  {stage:20s} +{latency:>6s} = {total:>5d}ms  {desc}')
print(f'\nTotal: ~{total}ms voice-to-voice')
print('\nOrchestration: LiveKit Agents or Pipecat')
```
