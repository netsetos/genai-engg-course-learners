## Ex 4: DreamFusion/SDS Conceptual


```python
print('Score Distillation Sampling (DreamFusion):')
print()
print('  1. Initialize random 3D representation (NeRF/mesh)')
print('  2. Render from random camera angle')
print('  3. Add noise to rendering')
print('  4. Pass to frozen 2D diffusion model:')
print('     "Does this look like [text prompt]?"')
print('  5. Compute gradient between predicted and actual noise')
print('  6. Backpropagate through differentiable renderer')
print('  7. Update 3D representation')
print('  8. Repeat from thousands of viewpoints')
print()
print('Result: 3D object that looks correct from EVERY angle')
print('Key insight: ZERO 3D training data needed!')
print()
print('2026: Feed-forward models (Meshy, Tripo) replaced SDS')
print('  SDS: hours per object (research)')
print('  Feed-forward: seconds per object (production)')
```
